# 1. Catalogue Effectiveness Holistically (/2)

#### Observation:
- Holistically, catalogues are driving an uplift on the two days that follow the drop. There is also a smaller uplift (5.2%) on Sundays during catalogue weeks.

#### Implication:

# 1. Catalogue Effectiveness Holistically (/3)

#### Observation:
- Holistically, catalogues are driving an uplift on the two days that follow the drop. There is also a smaller uplift (5.2%) on Sundays during catalogue weeks.

#### Implication:
- Catalogues are working primarily as a direct response medium (the biggest response is the day after the drop and the lion’s share of the response occurs within 2 days).

#### Proposed Actions:
- As catalogues are working as a DR medium, adopt the best-practices of DR marketing to drive results:
  - Only run catalogues with a very strong call to action (compelling offer).
  
##### Secondary Consideration:
- Running catalogues as close as possible to the biggest sales opportunity aligns to DR best practice. As Saturday is XXXXX sales peak, this suggests a Friday drop would be more effective than a Tuesday drop (noting that Salmatcurrently doesn’t accommodate Friday drops).
